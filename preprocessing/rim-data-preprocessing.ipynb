```python
import pandas as pd
import json
```

```python
data = pd.read_csv('data/CLEANED Usage (cleaned R13-08-20).csv')
data
```

```python
types = pd.Series(data['App Type Name'], dtype="category")
print(types.cat.categories)
```

```python
# getting dictionary, app type name as key and app name as value
# but this will give us repeated app names with the same type so we get the unique values for each key
app_types_names = data.groupby('App Type Name')['App Name'].apply(list).to_dict()
app_types_names_unique = {}
final_json = {}
for key, value in app_types_names.items():
    app_types_names_unique[key] = list(set(value))
for key, value in app_types_names_unique.items():
    to_add = {}
    items = []
    for item in value:
        to_add['name'] = item
        to_add['count'] = app_types_names[key].count(item)
        items.append(to_add)
        to_add = {}
        final_json[key] = items
```

```python
for i in final_json:
    print(len(final_json[i]), end=' ')
    print(i)
```

```python
# converting the dictionary to a new json file
# to use in the d3.js
with open("app_types_names_final.json", "w") as f:  
    json.dump(final_json, f)
```
